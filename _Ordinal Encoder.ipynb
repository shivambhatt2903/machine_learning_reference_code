### import libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
```

### import dataset

```python
df = pd.read_csv("ordinal.csv")
df.head()
```

### Feature Selection

```python
df=df.iloc[:,2:]
df.head()
```

### train_test_split

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=["purchased"]),df["purchased"],test_size=0.2,random_state=2)
```

### We use ordinal encoder for input variables 

```python
oe = OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
```

```python
x_train_oe=oe.fit_transform(x_train)
x_test_oe=oe.transform(x_test)
```

### We use labelencoder for target variable

```python
le = LabelEncoder()
```

```python
y_train_le = le.fit_transform(y_train)
y_test_le = le.transform(y_test)
```
