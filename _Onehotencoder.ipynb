### import libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

### import dataset

```python
df = pd.read_csv("one_hot_encoding.csv")
df.head()
```

### one hot encoding using pandas

```python
pd.get_dummies(df,columns=["fuel","owner"])
```

### using K-1 one hot encoding 

```python
pd.get_dummies(df,columns=["fuel","owner"],drop_first=True)
```

### train_test_split

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=["selling_price"]),df["selling_price"],test_size=0.3,random_state=3)
```

### One hot encoding using scikit learn

```python
ohe = OneHotEncoder(sparse_output=False,handle_unknown="ignore")
```

```python
x_train_fuel_owner = ohe.fit_transform(x_train[["fuel","owner"]])
x_test_fuel_owner = ohe.transform(x_test[["fuel","owner"]])
```

```python
x_train_fuel_owner
```

```python
x_train_fuel_owner.shape  # columns = 9
```

### One hot encoding K-1

```python
ohe2 = OneHotEncoder(sparse_output=False,handle_unknown="ignore",drop="first")
```

```python
x_train_ohe2 = ohe2.fit_transform(x_train[["fuel","owner"]])
x_test_ohe2 = ohe2.transform(x_test[["fuel","owner"]])
```
